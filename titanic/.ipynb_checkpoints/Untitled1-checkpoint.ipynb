```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df = train_df.drop(['PassengerId','Name','Cabin'], axis=1)
```

```python
train_df.info()
```

```python
# Pclass vs Survived
```

```python
train_df[['Pclass','Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending= False)
```

```python
train_df[['Sex','Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Sex', ascending= False)
```

```python
train_df.head()
```

```python
train_df = train_df.drop(['Ticket'], axis=1)
```

```python
train_df.head()
```

```python
# Converting to categorical vlaues
```

```python
combine = [train_df, test_df]
```

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map({'female':1, 'male':0}).astype(int)
    
```

```python
train_df.head()
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1

train_df[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_df.head()
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].dropna().map({'S':0,'C':1, 'Q':2}).astype(int)
    
```

```python
train_df.head()
```

```python
train_df.describe()
```

```python
test_df = test_df.drop(['PassengerId'], axis=1)
```

```python
test_df.describe()
```

```python
for dataset in combine:
    dataset['Age'].fillna(dataset['Age'].dropna().median(), inplace = True)
```
